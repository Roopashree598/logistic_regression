```python
#import pandas
import pandas as pd

#import dataset
cc_apps = pd.read_csv("cc_approvals.data", header=None)


# Inspect data
cc_apps.head()
```

**Inspecting the Applications**
The features of the dataset have been anonymized to protect the privacy.The probable features in a typical credit care application are gender,age,debt,Married,BankCustomer,EducationLevel, Ethnicity,YearsEmployed, PriorDefault, Employed, CreditScore, DriversLicense, Citizen, ZipCode, Income and finally the ApprovalStatus.We can map these features with respect to the columns in the output.

```python
# Print the statistics 
cc_apps_description=cc_apps.describe()
print(cc_apps_description)

print("\n")

#print DataFrame info
cc_apps_info=cc_apps.info()
print(cc_apps_info)

print("\n")

#Inspect mssing values in the dataset
cc_apps.tail(17)
```

**Handling the missing values**- In the above dataset they were miissing values which may yeild in accurate details.So in order to fill those missing values we first convert the missing values to NaN

```python
#import numpy
import numpy as np
 
#inspect missing values in the dataset
print(cc_apps.tail(17))

#replace the '?' with NaN
cc_apps= cc_apps.replace('?',np.nan)

#inspect the missing values again
cc_apps.tail(17)
```

So, now on converting all the missing values to NaN.By using the technique of mean imputation we will find those missing values.

```python
#impute the missing values with the mean imputation 
cc_apps.fillna(cc_apps.mean(),inplace=True)

#Count the number of NaNs in the dataset and print the counts to verify
print(cc_apps.isnull().sum())
```

Now all the missing values have been handled and it shows that no more missing values are present in dataset.But the non-numeric data yet has some missing values to fill those by using the technique by filling with the most frequent values as present in the respective columns




```python
#iterates over each column of cc_apps
for col in cc_apps.columns:
 
  #Checks if the column is of object type 
  if cc_apps[col].dtypes =='object':
    #impute with the most frequent value
    cc_apps=cc_apps.fillna(cc_apps[col].value_counts().index[0])

#count the number of NaNs in the dataset and print the counts to verify 
print(cc_apps.isnull().sum())
```

**Preprocessing the Data**
We need to split the data into train and test data and scale the values to uniform range.

Using the label encoding we are converting non-numeric features to numeric features using **label encoding**.


```python
#import LabelEncoder
from sklearn.preprocessing import LabelEncoder

#Instantiate LabelEncoder
le=LabelEncoder()

#Iterate over all the values of each column and extract their dtypes
for col in cc_apps.columns.to_numpy():
  #compare if dtype is object
  if cc_apps[col].dtypes=='object':
  #Use LabelEncoder to do the numeric transformation
    cc_apps[col]=le.fit_transform(cc_apps[col])
```

```python
print(cc_apps.info())

print("\n")

print(cc_apps.tail(17))
```

**Splitting of datasets**

Splitting of data into two diffrent phases of machine learning train and test.Test data is only used for validating it and not used for training.



```python
# Import train_test_split
from sklearn.model_selection import train_test_split 

#  convert the DataFrame to a NumPy array
cc_apps = cc_apps.to_numpy()

# Segregate features and labels into separate variables
X= cc_apps[:,13]  
y=cc_apps[:,13]


# Split into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X
                                                    ,
                                                    y,
                                                    test_size=0.33,
                                                    random_state=42)
```

**Preprocessing the Data**-Once we split the data into train and test.We have to scale the data before we fit a machine learning model to the data.

Like now we have a column called creditscore using this feature we can say the creditworthiness based on their credit history.More the creditscore shows he is more financially strong.Here the highest score is 1 since we are scaling the values in the range of [0-1].

```python
# Import MinMaxScaler
from sklearn.preprocessing import MinMaxScaler

# Instantiate MinMaxScaler and use it to rescale X_train and X_test
scaler = MinMaxScaler(feature_range=(0, 1))
rescaledX_train = scaler.fit_transform(X_train)
rescaledX_test = scaler.transform(X_test)
```

To achieve this model we use **Logistic regression**(a genearlized linear model) model to the train set.

Our main goal is to say if the credit card is approved or not for this we use **classification** task.

```python
# Import LogisticRegression
from sklearn.linear_model import LogisticRegression

# Instantiate a LogisticRegression classifier with default parameter values
logreg = LogisticRegression()

# Fit logreg to the train set
logreg.fit(rescaledX_train,y_train)
```

**Making predictions and evaluating performance**-We will have to evaluate the model on the test set with respect to **classification accuracy and confusion matrix**.

Its important to verify if the model is able to predict the approval status of the applications as denied that was originally that got denied.

Therfore,the confusion matrix helps us to view the model's performance from these aspects.




```python
# Import confusion_matrix
from sklearn.metrics import confusion_matrix

# Use logreg to predict instances from the test set and store it
y_pred = logreg.predict(rescaledX_test)

# Get the accuracy score of logreg model and print it
print("Accuracy of logistic regression classifier: ", logreg.score(rescaledX_test,y_test))

# Print the confusion matrix of the logreg model
confusion_matrix(y_test,y_pred)
```

**Grid searching and making the model perform better-**

The accuracy of model is 84%.To improve its performance we use grid search.


```python
# Import GridSearchCV
from sklearn.model_selection import GridSearchCV

# Define the grid of values for tol and max_iter
tol = [0.01, 0.001 ,0.0001]
max_iter = [100, 150, 200]

# Create a dictionary where tol and max_iter are keys and the lists of their values are the corresponding values
param_grid = dict(tol=tol, max_iter=max_iter)
```

**Finding the best performing model-**

We are defining the grid of hyperparameter values and converted them to a single dictionary format GridSearchCV()

We instantiate GridSearchCV() with our earlier logreg model with all the data we have. Instead of passing train and test sets separately, we will supply X (scaled version) and y. We will also instruct GridSearchCV() to perform a cross-validation of five folds.




```python
# Instantiate GridSearchCV with the required parameters
grid_model = GridSearchCV(estimator=logreg, param_grid=param_grid, cv=5)

# Use scaler to rescale X and assign it to rescaledX
rescaledX = scaler.fit_transform(X)

# Fit grid_model to the data
grid_model_result = grid_model.fit(rescaledX, y)

# Summarize results
best_score, best_params = grid_model_result.best_score_, grid_model_result.best_params_
print("Best: %f using %s" % (best_score, best_params))
```
